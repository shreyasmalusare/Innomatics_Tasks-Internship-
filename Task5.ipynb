#Problem 1-Social Media – Post Engagement Analyzer

```python
# Function to analyze post engagement
def analyze_engagement(likes_list):
    total_likes = 0
    
    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes
    
    # Conditional Logic
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    # Output
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example Data (likes from different hours)
likes_data = [200, 150, 300, 250, 150]

# Function Call
analyze_engagement(likes_data)
```

#Problem 3-Agriculture – Rainfall Adequacy Checker

```python
# Function to analyze rainfall data
def check_rainfall(rainfall_data, required_level):
    
    total_rainfall = 0
    
    # Loop to calculate total rainfall
    for rain in rainfall_data:
        total_rainfall += rain
    
    # Mathematical Operation (Average)
    average_rainfall = total_rainfall / len(rainfall_data)
    
    # Conditional Logic
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    # Output
    print("Average Rainfall:", int(average_rainfall))
    print("Rainfall Status:", status)


# Example Data (Seasonal rainfall in mm)
rainfall_list = [70, 80, 65, 75, 70]

# Required rainfall level
required = 70

# Function Call
check_rainfall(rainfall_list, required)
```

#Problem 4-Social Media – Duplicate Account Detection

```python
# Function to detect duplicate usernames
def detect_duplicate_accounts(usernames):
    
    # Convert list to set
    unique_users = set(usernames)
    
    # If length differs, duplicates exist
    if len(usernames) != len(unique_users):
        result = "Yes"
    else:
        result = "No"
    
    # Output
    print("Duplicate Accounts Found:", result)


# Example Data
user_list = ["john123", "emma45", "john123", "alex99"]

# Function Call
detect_duplicate_accounts(user_list)
```

#Problem 6-Agriculture – Premium Crop Price Filter

```python
# Function to filter premium crop prices
def filter_premium_crops(price_list):
    
    premium_crops = []
    
    # Loop for filtering
    for price in price_list:
        if price > 2000:
            premium_crops.append(price)
    
    # Output
    print("Premium Crops:", premium_crops)


# Example Data
crop_prices = [1500, 2500, 1800, 3200, 1900]

# Function Call
filter_premium_crops(crop_prices)
```

#Problem 9: E-Learning – Student Attendance Eligibility

```python
# Function to calculate attendance and check eligibility
def check_exam_eligibility(attendance_list):
    
    total_classes = len(attendance_list)
    attended_classes = 0
    
    # Loop to count attended days
    for day in attendance_list:
        if day == 1:   # 1 means Present, 0 means Absent
            attended_classes += 1
    
    # Calculate attendance percentage
    attendance_percentage = (attended_classes / total_classes) * 100
    
    # Conditional Check
    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    # Output
    print("Attendance Percentage:", attendance_percentage)
    print("Exam Eligibility:", status)


# Example Data (1 = Present, 0 = Absent)
attendance_data = [1,1,1,0,1,1,1,1,0,1]   # 8/10 = 80%

# Function Call
check_exam_eligibility(attendance_data)
```

#Problem 10: Smart Electricity Bill Analyzer

```python
# Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    
    bill = 0
    
    # Slab Calculation Logic
    if units <= 100:
        bill = units * 3
        
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
        
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    
    # Usage Classification Logic
    if bill < 500:
        usage_status = "Low Usage"
    elif 500 <= bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
    
    
    # Return Multiple Values
    return bill, usage_status


# Example Usage
units_consumed = 250

total_bill, status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Classification:", status)
```
