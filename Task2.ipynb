```python
##Task 3-Find Maximum and Minimum Values From a List of Numbers
```

```python
# We have taken a list of numbers
numbers = [45, 22, 89, 10, 66]

# Initialize maximum and minimum with the first element of the list
maximum = numbers[0]
minimum = numbers[0]

# for loop to traverse each element in the list
for num in numbers:
    
    # check if current number is greater than maximum
    if num > maximum:
        maximum = num
    
    # check if current number is smaller than minimum
    if num < minimum:
        minimum = num

# print the original list
print("List:", numbers)

# print maximum and minimum values
print("Max:", maximum)
print("Min:", minimum)
```

```python
##Task 4-Count Products Above a Price Threshold
```

```python
# We have taken a list of product prices
prices = [450, 1200, 899, 1500, 300]

# Threshold price value
threshold = 1000

# Variable to count products above the threshold
count = 0

# for loop to iterate through each price in the list
for price in prices:
    
    # check if the price is greater than 1000
    if price > threshold:
        count = count + 1   # increment count

# print the result
print("Products above 1000:", count)
```

```python
#Task 5-Calculating Of Attendance Percentage
```

```python
# We have taken a list representing attendance
# 'P' means Present and 'A' means Absent
attendance = ["P", "P", "A", "P", "P"]

# Variable to count number of present days
present_count = 0

# Total number of days
total_days = len(attendance)

# for loop to check each attendance entry
for status in attendance:
    
    # check if the student is present
    if status == "P":
        present_count = present_count + 1

# calculate attendance percentage
attendance_percentage = (present_count / total_days) * 100

# print the result
print("Attendance Percentage:", attendance_percentage)
```

```python
#Task 7-Count Character Frequency
```

```python
# We have taken a string input
text = "pythonp"

# Create an empty dictionary to store character frequencies
char_count = {}

# for loop to iterate through each character in the string
for char in text:
    
    # if character already exists in dictionary
    if char in char_count:
        char_count[char] = char_count[char] + 1
    else:
        # if character does not exist, initialize count as 1
        char_count[char] = 1

# print the character frequency dictionary
print(char_count)
```

```python
##Task 10-Calculate Average Marks
```

```python
# We have taken a list of marks obtained by the student
marks = [70, 75, 80, 65]

# Variable to store the total sum of marks
total_marks = 0

# Total number of subjects
total_subjects = len(marks)

# for loop to calculate the sum of marks
for mark in marks:
    total_marks = total_marks + mark

# calculate average marks
average_marks = total_marks / total_subjects

# print the result
print("Average Marks:", average_marks)
```
