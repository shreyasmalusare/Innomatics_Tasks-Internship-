```python
#Task-2
#Pass-Fail Analyzer
```

```python
#Pass-Fail analyzer
marks = [45, 78, 90, 33, 60]   #Taken array of marks

pass_count = 0   #Initially pass count
fail_count = 0   #Initially fail-count

for mark in marks:    #for loop for traversing marks in the array
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Pass Students:", pass_count)  #print pass students coun
print("Total Fail Students:", fail_count)  #print fail students count
```

```python
#Task-3
#Simple Data Cleaner
```

```python
# Create a list of names with extra spaces and inconsistent letter casing
names = [" Alice ", "bob", " CHARLIE "]

# Create an empty list to store the cleaned names
cleaned_names = []

# Loop through each name in the names list
for name in names:
    
    # Remove leading and trailing spaces from the name
    # Convert the name to lowercase
    # Add the cleaned name to the cleaned_names list
    cleaned_names.append(name.strip().lower())

# Print the final cleaned list of names
print("Cleaned Names:", cleaned_names)
```

```python
#Task-4
#Message Length Analyzer
```

```python
# Message Length Analyzer
# This program analyzes the length of each message in the list

# List of messages to be analyzed
messages = ["Hi", "Welcome to the platform", "OK"]

# Loop through each message in the list
for message in messages:
    
    # Calculate the length of the current message
    length = len(message)
    
    # Print the message along with its length
    print(f"Message: '{message}' , Length: {length}")
    
    # Check if the message length is greater than 10 characters
    if length > 10:
        
        # Print a warning message for long messages
        print("Message longer than 10 characters")
```

```python
# Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR messages:", error_count)
```
