```python
# Smart Light Controller Program
# This program is used to control a light (ON/OFF)

class SmartLight:
    
    # Constructor to store light name
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"   # Initially light is OFF
    
    # Method to turn light ON or OFF
    def control_light(self, action):
        
        if action == "ON":
            self.status = "ON"
        
        elif action == "OFF":
            self.status = "OFF"
        
        else:
            print("Invalid Action!")
            return
        
        print(self.light_name, "is", self.status)


# Creating object
light1 = SmartLight("Bedroom Light")

# Performing action
light1.control_light("ON")
```

```python
# Employee ID Card System
# This program stores employee details and displays ID card

class Employee:
    
    # Constructor to store employee details
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    
    # Method to display ID card
    def display_id_card(self):
        
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Creating object
emp1 = Employee("Rahul", "EMP102", "IT")

# Displaying ID card
emp1.display_id_card()
```

```python
# Mobile Contact Record Program
# This program stores contact details

class Contact:
    
    # Constructor to store contact name and phone number
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    
    # Method to display contact information
    def display_contact(self):
        
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Creating object
contact1 = Contact("Anita", "9876543210")

# Displaying contact details
contact1.display_contact()
```

```python
# Product Price Tag Generator
# This program prints product name and price

class Product:
    
    # Constructor to store product details
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price
    
    # Method to display price tag
    def display_price_tag(self):
        
        print("Product:", self.product_name)
        print("Price: ₹" + str(self.price))


# Creating object
product1 = Product("Headphones", 2499)

# Displaying price tag
product1.display_price_tag()
```

```python
# Movie Rating Display System
# This program stores movie details and rating

class Movie:
    
    # Constructor to store movie details
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating
    
    # Method to display movie details
    def display_movie(self):
        
        print("Movie:", self.movie_name)
        print("Rating:", self.rating, "/ 5")


# Creating object
movie1 = Movie("Inception", 4.8)

# Displaying movie details
movie1.display_movie()
```

```python
# Delivery Address Manager
# This program stores customer delivery details

class Delivery:
    
    # Constructor to store customer name and address
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    
    # Method to display delivery details
    def display_details(self):
        
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Creating object
delivery1 = Delivery("Suman", "Hyderabad")

# Displaying delivery details
delivery1.display_details()
```
