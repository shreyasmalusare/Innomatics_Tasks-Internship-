#Problem-1:Smart Parking Lot Management System

```python
#Problem1- Function to manage smart parking system
def smart_parking_system(capacity, logs):

    # Variable to store current number of parked vehicles
    current_parked = 0

    # Variable to track maximum vehicles parked at any time (peak usage)
    peak_usage = 0

    # Loop through each log entry in the logs list
    for log in logs:

        # If a vehicle enters, increase parked count
        if log == "IN":
            current_parked += 1   # Increment by 1 for each entry

        # If a vehicle exits, decrease parked count
        elif log == "OUT":

            # Check to avoid negative parking count
            if current_parked > 0:
                current_parked -= 1   # Decrement by 1 for exit

        # After each operation, check if this is the highest usage so far
        if current_parked > peak_usage:
            peak_usage = current_parked   # Update peak usage

    # Display final number of parked vehicles
    print("Currently Parked Vehicles:", current_parked)

    # Display peak parking usage
    print("Peak Parking Usage:", peak_usage)

    # Check if parking exceeds capacity
    if current_parked > capacity:
        print("Parking Status: Full - Capacity Exceeded!")  # Alert message
    else:
        print("Parking Status: Available")   # Parking available


# Given parking capacity
capacity = 50

# Given vehicle logs
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

# Function call to execute parking system
smart_parking_system(capacity, logs)
```

#Problem 2:Online Food Delivery Time Estimator

```python
# Function to estimate delivery time
def estimate_delivery_time(distance, traffic, weather):

    # Step 1: Calculate base delivery time
    # Assuming 1 km takes 5 minutes
    base_time = distance * 5

    # Initialize total time with base time
    total_time = base_time

    # Step 2: Add delay based on traffic condition
    if traffic == "High":
        total_time += 15   # Add 15 minutes for high traffic
    elif traffic == "Medium":
        total_time += 10   # Add 10 minutes for medium traffic
    elif traffic == "Low":
        total_time += 5    # Add 5 minutes for low traffic

    # Step 3: Add delay based on weather condition
    if weather == "Rainy":
        total_time += 10   # Add 10 minutes for rain
    elif weather == "Storm":
        total_time += 20   # Add 20 minutes for storm
    elif weather == "Clear":
        total_time += 0    # No delay for clear weather

    # Step 4: Display final estimated delivery time
    print("Estimated Delivery Time:", total_time, "minutes")


# Input
distance = 8
traffic = "High"
weather = "Rainy"

# Function call
estimate_delivery_time(distance, traffic, weather)
```

```python
# Function to analyze movie theatre seat occupancy
def movie_seat_analyzer(total_seats, booked_seats):

    # Count total number of booked seats using length of list
    total_booked = len(booked_seats)

    # Calculate occupancy percentage
    occupancy_percentage = (total_booked / total_seats) * 100

    # Print occupancy percentage (rounded for neat output)
    print("Occupancy:", round(occupancy_percentage), "%")

    # Check if show is completely full
    if occupancy_percentage == 100:
        print("Show Status: Housefull")

    # If occupancy is between 70% and 99%
    elif occupancy_percentage >= 70:
        print("Show Status: Almost Full")

    # If occupancy is below 70%
    else:
        print("Show Status: Seats Available")

    # Suggest additional shows if occupancy is very high
    if occupancy_percentage >= 90:
        print("Suggestion: Consider Opening Additional Shows")


# Sample Input
total_seats = 200

# Simulating 150 booked seats (150 entries of 1)
booked_seats = [1] * 150

# Function call
movie_seat_analyzer(total_seats, booked_seats)
```

#Problem 4-Cloud Server Load Classification System

```python
# Function to classify cloud server load
def classify_server_load(cpu_readings):

    # Variable to store total CPU usage
    total = 0

    # Loop through each CPU reading in the list
    for reading in cpu_readings:
        total += reading   # Add each reading to total

    # Calculate average CPU usage
    average = total / len(cpu_readings)

    # Print average CPU load (rounded for clean output)
    print("Average CPU Load:", round(average), "%")

    # Classify server status based on average
    if average < 50:
        print("Server Status: Normal")

    elif average <= 80:
        print("Server Status: Warning")

    else:
        print("Server Status: Critical")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

# Function call
classify_server_load(cpu_readings)
```

#Problem 5-Smart Classroom Resource Usage Monitor

```python
# Function to monitor classroom resource usage
def classroom_resource_monitor(resource_usage):

    # Define maximum allowed usage hours per day
    usage_limit = 8

    # Create an empty list to store overused resources
    overused_resources = []

    # Loop through each resource and its usage hours
    for resource, hours in resource_usage.items():

        # Check if usage exceeds the allowed limit
        if hours > usage_limit:
            overused_resources.append(resource)   # Add resource to overused list

    # Display overused resources
    if len(overused_resources) > 0:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# Function call
classroom_resource_monitor(resource_usage)
```

#Problem 6-Online Event Registration Capacity Controller

```python
# Function to manage event registrations
def event_registration_controller(capacity, total_registrations):

    # Variable to count confirmed registrations
    confirmed = 0

    # Variable to count waitlisted users
    waitlisted = 0

    # Loop through each registration attempt
    for i in range(total_registrations):

        # If confirmed registrations are less than capacity
        if confirmed < capacity:
            confirmed += 1   # Approve registration

        # If capacity is full
        else:
            waitlisted += 1   # Add user to waitlist

    # Print confirmed registrations
    print("Confirmed Registrations:", confirmed)

    # Print waitlisted users
    print("Waitlisted Users:", waitlisted)

    # Determine registration status
    if confirmed == capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Sample Input
capacity = 100
total_registrations = 105

# Function call
event_registration_controller(capacity, total_registrations)
```
