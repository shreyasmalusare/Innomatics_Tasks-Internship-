#Task 3-Employee Performance Bonus Eligibility

```python
# Create a dictionary to store employee names as keys
# and their performance scores as values
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest performance score
# The max() function is used to get the highest value
# from the dictionary's values (scores)
highest_score = max(employees.values())

# Step 2: Find all employees who have the highest score
# We use a list comprehension to loop through each
# (name, score) pair in the dictionary using items()
# If the employee's score matches the highest_score,
# we add their name to the top_performers list
top_performers = [name for name, score in employees.items() if score == highest_score]

# Step 3: Convert the list of top performers into a string
# The join() method combines the names into a single string
# separated by commas
top_performers_string = ", ".join(top_performers)

# Step 4: Display the final result
# We use an f-string to format the output clearly
print(f"Top Performers Eligible for Bonus: {top_performers_string} (Score: {highest_score})")
```

#Problem 2-An e-commerce website stores customer search queries.

```python
# Import the string module
# This module contains a list of punctuation characters
import string

# Store the search query entered by the user
query = "Buy mobile phone buy phone online"

# Step 1: Convert the entire sentence to lowercase
# This ensures that 'Buy' and 'buy' are treated as the same word
query = query.lower()

# Step 2: Remove punctuation from the sentence
# string.punctuation contains characters like !@#$%^&*() etc.
# We loop through each punctuation character and remove it
for char in string.punctuation:
    query = query.replace(char, "")

# Step 3: Split the sentence into individual words
# The split() function separates words based on spaces
words = query.split()

# Step 4: Create an empty dictionary to store word frequencies
frequency = {}

# Step 5: Count how many times each word appears
# We loop through each word in the list
for word in words:
    
    # If the word is already in the dictionary,
    # increase its count by 1
    if word in frequency:
        frequency[word] += 1
    
    # If the word is not in the dictionary,
    # add it with count 1
    else:
        frequency[word] = 1

# Step 6: Create a new dictionary to store
# only words that appear more than once
result = {}

# Loop through the frequency dictionary
for word, count in frequency.items():
    
    # Check if the word count is greater than 1
    if count > 1:
        result[word] = count

# Step 7: Display the final result
print(result)
```

#Problem 3-SENSOR DATA VALIDATION

```python
# Store the sensor readings in a list
# The index represents the hour
# The value at each index represents the sensor reading at that hour
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Create an empty list to store valid readings
# Each valid reading will be stored as a tuple:
# (hour_index, reading_value)
valid_readings = []

# Use a loop to go through the list
# enumerate() gives both:
# 1. index (hour)
# 2. value (sensor reading)
for hour, value in enumerate(sensor_readings):
    
    # Check if the reading is even
    # A number is even if dividing by 2 leaves remainder 0
    if value % 2 == 0:
        
        # If the reading is even, add a tuple
        # containing (hour, value) to valid_readings list
        valid_readings.append((hour, value))

# Print the final list of valid sensor readings
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

#Problem 4-EMAIL DOMAIN USAGE ANALYSIS

```python
# Store the list of employee email IDs
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Step 1: Create an empty dictionary
# This will store domain names as keys
# and the count of users as values
domain_count = {}

# Step 2: Loop through each email in the list
for email in emails:
    
    # Step 3: Extract the domain part
    # Split the email using '@'
    # The domain is the second part (index 1)
    domain = email.split("@")[1]
    
    # Step 4: Count occurrences of each domain
    # If domain already exists in dictionary,
    # increase its count by 1
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        # If domain does not exist, add it with count 1
        domain_count[domain] = 1

# Step 5: Calculate total number of emails
total_emails = len(emails)

# Step 6: Calculate percentage usage for each domain
# Loop through the dictionary items (domain and count)
for domain, count in domain_count.items():
    
    # Percentage formula:
    # (domain count / total emails) * 100
    percentage = (count / total_emails) * 100
    
    # Print domain with percentage
    # :.0f formats the percentage without decimal places
    print(f"{domain}: {percentage:.0f}%")
```

#Problem 5-SALES SPIKE DETECTION

```python
# Store daily sales values in a list
# Each index represents a day (index 0 = Day 1)
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate the total sales
# sum() adds all values in the list
total_sales = sum(sales)

# Step 2: Calculate the number of days
# len() returns the number of elements in the list
number_of_days = len(sales)

# Step 3: Calculate the average daily sales
# Average formula = Total Sales / Number of Days
average_sales = total_sales / number_of_days

# Step 4: Calculate the spike threshold
# A spike is defined as more than 30% above average
# So we multiply average by 1.30 (which is 100% + 30%)
threshold = average_sales * 1.30

# Step 5: Loop through sales list with index and value
# enumerate() gives both day index and sales amount
for index, value in enumerate(sales):
    
    # Day number should start from 1 instead of 0
    day_number = index + 1
    
    # Step 6: Check if the sale is greater than threshold
    if value > average_sales:
        
        # If condition is true, print the day and sale value
        print(f"Day {day_number}: {value}")
```

#Prolem 6-DUPLICATE USER-ID DETECTION

```python
# Store user IDs in a list
# Some IDs are repeated (duplicates)
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Step 1: Create an empty dictionary
# This dictionary will store:
# Key   → user ID
# Value → number of times it appears
frequency = {}

# Step 2: Count how many times each user ID appears
for user in user_ids:
    
    # If user ID already exists in dictionary,
    # increase its count by 1
    if user in frequency:
        frequency[user] += 1
    
    # If user ID is not in dictionary,
    # add it with count 1
    else:
        frequency[user] = 1

# Step 3: Display only duplicate IDs
# A duplicate means count > 1
for user, count in frequency.items():
    
    # Check if the ID appears more than once
    if count > 1:
        
        # Print user ID and number of occurrences
        print(f"{user} → {count} times")
```
