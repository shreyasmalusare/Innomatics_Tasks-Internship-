#Task 4-PROBLEM 1-MOBILE RECHARGE VALIDATION SYSTEM

```python
#Problem 1-Mobile Recharge Validation System
```

```python
# Define a function to validate recharge amount
def validate_recharge(amount):
    """
    This function checks whether the recharge amount is valid.
    
    Rules:
    1. Amount must be greater than or equal to 50
    2. Amount must match one of the valid recharge plans
    """

    # List of valid recharge plans
    valid_plans = [199, 299, 399, 599]

    # Rule 1: Minimum recharge check
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False

    # Rule 2: Check if amount is one of the valid plans
    if amount not in valid_plans:
        print("Invalid recharge plan.")
        print("Available plans are:", valid_plans)
        return False

    # If both conditions pass
    return True


# Main program starts here

# Infinite loop to allow user retry
while True:
    
    # Take recharge amount input from user
    # Convert input string into integer
    amount = int(input("Enter recharge amount: ₹"))

    # Call the validation function
    if validate_recharge(amount):
        
        # If function returns True, recharge is successful
        print("Recharge Successful ✅")
        
        # Break the loop (stop retrying)
        break
    else:
        # If invalid, allow user to retry
        print("Please try again.\n")
```

#Problem 2-INVENTORY REORDER ALERT SYSTEM

```python
# Define a function to check inventory levels
def check_inventory(products):
    """
    This function checks stock quantity for each product.
    
    If stock < 15  → Print 'Reorder Alert'
    Otherwise      → Print 'Stock OK'
    """

    # Loop through dictionary using items()
    # product → key (product name)
    # stock   → value (quantity available)
    for product, stock in products.items():
        
        # Check if stock is less than 15
        if stock < 15:
            
            # If true, print reorder alert
            print(f"{product}: Reorder Alert (Stock: {stock})")
        
        else:
            # Otherwise, stock level is fine
            print(f"{product}: Stock OK (Stock: {stock})")


# Sample inventory dictionary
inventory = {
    "Laptop": 20,
    "Mouse": 10,
    "Keyboard": 5,
    "Monitor": 18
}

# Call the function and pass the dictionary
check_inventory(inventory)
```

#Problem 3-Student Result Processing System

```python
# Define a function to process student result
def process_student_result(marks):
    """
    This function accepts a list of marks,
    calculates average, and returns 'Pass' or 'Fail'.
    
    Rules:
    - Average >= 50 → Pass
    - Average < 50 → Fail
    """
    
    # Step 1: Initialize total marks
    total = 0
    
    # Step 2: Loop through the list to sum the marks
    for mark in marks:
        total += mark
    
    # Step 3: Calculate average
    average = total / len(marks)
    
    # Step 4: Check if average is ≥ 50
    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Sample list of student marks
student_marks = [60, 45, 70, 50]

# Call the function and store result
result = process_student_result(student_marks)

# Print the result
print("Student Result:", result)
```

#Problem 4-CAB FARE ESTIMATOR WITH RETRY OPTION

```python
# Define a function to calculate cab fare
def calculate_fare(distance_km, peak_hour=False):
    """
    Calculates cab fare based on:
    - Base fare = ₹50
    - ₹12 per km
    - 25% extra during peak hours
    """
    
    base_fare = 50
    per_km_rate = 12

    # Step 1: Calculate initial fare
    fare = base_fare + (distance_km * per_km_rate)

    # Step 2: Add peak hour surcharge if applicable
    if peak_hour:
        fare += fare * 0.25  # 25% extra

    return fare


# Main program with retry option
while True:
    try:
        # Step 3: Take user input for distance
        distance = float(input("Enter distance in km: "))

        # Step 4: Ask if it's peak hour
        peak_input = input("Is it peak hour? (yes/no): ").strip().lower()
        peak = True if peak_input == "yes" else False

        # Step 5: Calculate fare using function
        total_fare = calculate_fare(distance, peak)

        # Step 6: Display total fare
        print(f"Total Fare: ₹{total_fare:.2f}")

        # Step 7: Break the loop after successful calculation
        break

    except ValueError:
        # Handle invalid input (non-numeric distance)
        print("Invalid input! Please enter numeric distance.\n")
        # Loop continues, allowing retry
```

#Problem 5-EMPLOYEE ATTENDANCE ELIGIBILITY CHECKER

```python
# Define a function to check attendance eligibility
def check_attendance(attendance_list):
    """
    This function calculates attendance percentage and
    returns 'Eligible' if attendance >= 75%, else 'Not Eligible'.

    Parameters:
    - attendance_list : list of "P" (Present) and "A" (Absent)

    Returns:
    - 'Eligible' or 'Not Eligible'
    """

    # Step 1: Initialize counter for present days
    present_days = 0

    # Step 2: Loop through the attendance list
    for day in attendance_list:
        if day.upper() == "P":  # Count only present days
            present_days += 1

    # Step 3: Calculate attendance percentage
    total_days = len(attendance_list)
    attendance_percentage = (present_days / total_days) * 100

    # Step 4: Check eligibility
    if attendance_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Sample attendance list
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

# Call the function and store result
result = check_attendance(attendance)

# Print the eligibility
print("Employee Attendance Eligibility:", result)
```

#Problem 6-PASSWORD STRENGTH CHECKER

```python
# Define a function to check password strength
def check_password_strength(password):
    """
    This function checks if a password is strong based on:
    1. Minimum length = 8 characters
    2. Must contain at least one digit
    3. Must contain at least one special character (@, #, $)
    
    Returns 'Strong Password' if all conditions met,
    else returns 'Weak Password'
    """

    # Step 1: Check minimum length
    if len(password) < 8:
        return "Weak Password: Must be at least 8 characters long"

    # Step 2: Initialize flags for digit and special character
    has_digit = False
    has_special = False

    # Step 3: Loop through each character in password
    for char in password:
        if char.isdigit():
            has_digit = True  # Found a digit
        elif char in "@#$":
            has_special = True  # Found a special character

    # Step 4: Check if both digit and special character exist
    if has_digit and has_special:
        return "Strong Password ✅"
    else:
        return "Weak Password: Must contain at least one digit and one special character (@, #, $)"


# Main program to test the function
password_input = input("Enter your password: ")

# Call the function and store the result
strength = check_password_strength(password_input)

# Display the password strength
print(strength)
```
